# Importing data

```python
# Importing the initial dataset.
df = pd.read_csv("clean_data.csv", encoding="ISO-8859-1")

# Extracting all the bookID's as string to a list from the pandas dataframe.
bookID_list = []
for i in range(len(df["bookID"])):
    bookID_list.append(str(df["bookID"][i]))

# Creating a URL list with a for loop. Concatenating url of goodreads with bookID's.
url_list = []
for i in range(len(bookID_list)):
    url_list.append("https://www.goodreads.com/book/show/" + bookID_list[i])
```

# Scraping Genres

```python
# The code for scraping genres of a book from goodreads, takes a list of urls, list of book id's and number of genres as paramaters.
# Modified it into a function.

def extract_book_genres(url_list, bookID_list, num_genres):

    y=0

    genres_dict={}

    for url in url_list:

        # sleep(randint(1,2))

        try:
            result = requests.get(url)
            # rest of code here
        except requests.exceptions.RequestException as e:
            # handle error here, e.g. log it and continue with the next URL
            print(f"Error fetching {url}: {e}")
            continue

        doc = BeautifulSoup(result.text, "html.parser")

        # find all a tags with class "Button--tag-inline"
        a_tags = doc.find_all('a', {'class': 'Button--tag-inline'})

        # Loop over the genres on a book. Add them to genres_of_a_book list. After getting 5 genres of the book, add it to the book_genres_dict. Key as the bookID and value as the genres_of_a_book.
        genres_of_a_book = []
        
        bookID = bookID_list[y]
        i=0
        for a_tag in a_tags:
            span_tag = a_tag.find('span', {'class': 'Button__labelItem'})
            genre = span_tag.text.strip()
            genres_of_a_book.append(genre)
            i+=1
            if i==num_genres:
                genres_dict[bookID] = genres_of_a_book
                break
        
        y+=1

        # if len(genres_dict)==3:
        #     break

    return genres_dict
```

```python
# Function to scrape the description of the books.

def description_scraping(url_list, bookID_list):

    book_description_dict = {}

    y=0

    for url in url_list:

        result = requests.get(url)
        doc = BeautifulSoup(result.text, "html.parser")

        book_id=bookID_list[y]

        span_tag = doc.find('span', {'class': 'Formatted'})

        if span_tag is not None:
            text = span_tag.text.strip()
            book_description_dict[book_id] = text
        else:
            book_description_dict[book_id] = "blank"

        y+=1

        # if len(book_description_dict) > 500:
        #     break
    
    return book_description_dict
```

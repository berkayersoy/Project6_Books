```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
import time
import pickle
```

```python
# Importing the initial dataset.
df = pd.read_csv("books.csv", encoding="ISO-8859-1")

# Extracting all the bookID's as string to a list from the pandas dataframe.
bookID_list = []
for i in range(len(df["bookID"])):
    bookID_list.append(str(df["bookID"][i]))

# Creating a URL list with a for loop. Concatenating url of goodreads with bookID's.
url_list = []
for i in range(len(bookID_list)):
    url_list.append("https://www.goodreads.com/book/show/" + bookID_list[i])
```

```python
# THE CODE THAT WORKES FOR 5 GENRES

book_genres_dict2 = {}

y=0

for url in url_list:
    result = requests.get(url)
    doc = BeautifulSoup(result.text, "html.parser")

    # find all a tags with class "Button--tag-inline"
    a_tags = doc.find_all('a', {'class': 'Button--tag-inline'})

    # Loop over the genres on a book. Add them to genres_of_a_book list. After getting 5 genres of the book, add it to the book_genres_dict. Key as the bookID and value as the genres_of_a_book.
    genres_of_a_book = []
    
    bookID = bookID_list[y]
    i=0
    for a_tag in a_tags:
        span_tag = a_tag.find('span', {'class': 'Button__labelItem'})
        genre = span_tag.text.strip()
        genres_of_a_book.append(genre)
        i+=1
        if i==3:
            book_genres_dict2[bookID] = genres_of_a_book
            break
    y+=1

print(book_genres_dict2)
```

```python
# MODIFIED VERSION WITH TRY EXCEPT FOR REQUESTS
# LETS TRY TO SCRAPE THE REST OF THE BOOKS FOR 3 GENRES


def extract_book_genres(url_list, bookID_list, num_genres):
    book_genres_dict = {}
    y = 0

    x=0

    for url in url_list:
        try:
            result = requests.get(url)
            doc = BeautifulSoup(result.text, "html.parser")

            # find all a tags with class "Button--tag-inline"
            a_tags = doc.find_all('a', {'class': 'Button--tag-inline'})

            # Filter out non-genre values
            genres_of_a_book = []
            for a_tag in a_tags:
                span_tag = a_tag.find('span', {'class': 'Button__labelItem'})
                genre = span_tag.text.strip()
                if not any(c.isnumeric() for c in genre):
                    genres_of_a_book.append(genre)
                if len(genres_of_a_book) == num_genres:
                    break
            
            # Add None values to genres_of_a_book to fill in missing genres
            while len(genres_of_a_book) < num_genres:
                genres_of_a_book.append(None)
            
            bookID = bookID_list[y]
            book_genres_dict[bookID] = genres_of_a_book
            
            y += 1

            time.sleep(1)
            
        except requests.exceptions.RequestException as e:
            print(f"An error occurred while processing URL {url}: {e}")
            continue
    
        x+=1
        if x==5:
            break

    return book_genres_dict

extract_genres_test=extract_book_genres(url_list2, bookID_list2, 3)
print(extract_genres_test)
```

```python
# THE CODE THAT WORKES FOR 5 GENRES
# Modified it into a function.

def extract_book_genres(url_list, bookID_list, num_genres):

    y=0

    genres_dict={}

    for url in url_list:


        try:
            result = requests.get(url)
            # rest of code here
        except requests.exceptions.RequestException as e:
            # handle error here, e.g. log it and continue with the next URL
            print(f"Error fetching {url}: {e}")
            continue

        doc = BeautifulSoup(result.text, "html.parser")

        # find all a tags with class "Button--tag-inline"
        a_tags = doc.find_all('a', {'class': 'Button--tag-inline'})

        # Loop over the genres on a book. Add them to genres_of_a_book list. After getting 5 genres of the book, add it to the book_genres_dict. Key as the bookID and value as the genres_of_a_book.
        genres_of_a_book = []
        
        bookID = bookID_list[y]
        i=0
        for a_tag in a_tags:
            span_tag = a_tag.find('span', {'class': 'Button__labelItem'})
            genre = span_tag.text.strip()
            genres_of_a_book.append(genre)
            i+=1
            if i==num_genres:
                genres_dict[bookID] = genres_of_a_book
                break
        
        y+=1

    return genres_dict

extract2 = extract_book_genres(url_list2, bookID_list2, 3)
```
